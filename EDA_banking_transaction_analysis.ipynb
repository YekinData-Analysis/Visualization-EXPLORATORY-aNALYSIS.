```python
#let run the maximum and the minimum:


def minmax(X):
    mini=ndf[X].idxmin()
    maxi=ndf[X].idxmax()
    maxii=pd.DataFrame(ndf.loc[maxi,:])
    minii=pd.DataFrame(ndf.loc[mini,:])
    return pd.concat([maxii,minii],axis=1)
```
