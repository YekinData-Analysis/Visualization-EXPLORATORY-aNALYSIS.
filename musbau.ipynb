```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Musbau(N=1000, rseed=42):
    # Making a plot that will carry my Own name "HELLO" as text;
    fig, ax = plt.subplots(figsize=(4, 1))
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
    ax.axis('off')
    ax.text(0.5, 0.7, 'MUSBAU', va='center', ha='center', weight='bold', size=60)
    fig.savefig('Musbau.png')
    plt.close(fig)
    
    # Open this PNG and draw random points from it
    from matplotlib.image import imread
    df = imread('Musbau.png')[::-1, :, 0].T
    rng = np.random.RandomState(rseed)
    X = rng.rand(4 * N, 2)
    i, j = (X * df.shape).astype(int).T
    mask = (df[i, j] < 1)
    X = X[mask]
    X[:, 0] *= (df.shape[0] / df.shape[1])
    X = X[:N]
    return X[np.argsort(X[:, 0])]
```

```python
X = Musbau(1000)
colorize = dict(c=X[:, 0], cmap=plt.cm.get_cmap('rainbow', 9))
plt.scatter(X[:, 0], X[:, 1], **colorize)
plt.axis('equal');
```
